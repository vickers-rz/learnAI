**对比实验展示生成器函数的惰性产出特性**

```
list_1 = list(range(1000000000000001)) # 创建一个大列表,会超出内存限制：运行时会抛出内存溢出错误：MemoryError
for i in list_1:
    print(i)
```

```
# 2. 创建生成器函数，逐个产生结果
def generator_function(n):
    for i in range(n):
        print(f"计算并产出第{i}个元素")
        yield i ** 2

for i in generator_function(2):
    print(i)
```

## 生成器与普通函数的区别

通过以上实验可以清楚地看出：

1. **普通函数**：
   - 立即执行所有计算
   - 一次性返回完整结果
   - 占用较多内存存储所有结果
   
2. **生成器函数**：
   - 惰性求值，在需要时才进行计算
   - 逐个产生结果
   - 内存效率高，只在需要时占用内存
   - 可以表示无限序列

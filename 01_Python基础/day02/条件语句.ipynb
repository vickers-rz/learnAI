```python
# 1. 输入三个值，分别表示箱子长宽高，然后输出箱子是长方体还是正方体
# input会将输入的数据全部转化为字符串
x = int(input("请输入箱子的长度："))
y = int(input("请输入箱子的宽度："))
z = int(input("请输入箱子的高度："))
if x == y and y == z:
    print("箱子是正方体")
else:
    print("箱子是长方体")
```

```python
# 2.输入学生的考试成绩，判断等级并输出
score = int(input("请输入学生的考试成绩："))

# 满分100
# 成绩大于90包括90的，A
# 成绩大于80包括80的，B
# 成绩大于70包括70的，C
# 成绩大于60包括60的，D
# 成绩小于60的   不合格

# 如果score大于等于90  并且  小于等于100 
if score >= 90 and score <= 100:
    print("A")
elif score >= 80 and score < 90:
    print("B")
elif score >= 70 and score < 80:
    print("C")
elif score >= 60 and score < 70:
    print("D")
elif score >= 0 and score < 60:
    print("成绩不合格！！！")
else:
    print("输入的成绩不符合要求")
```

```python
# 3. 输入三角形的三条边长，判断是否能够构成三角形，如果能，输出是什么类型的三角形
# 等腰  等边  直角  普通
# 如果不能构成三角形，输出不能构成三角形

a = int(input("请输入三角形的第一条边长："))
b = int(input("请输入三角形的第二条边长："))
c = int(input("请输入三角形的第三条边长："))

# 三角形的构成条件：任意两条边的边长加起来大于第三边
if a + b > c and a + c > b and b + c > a:
    # 等腰：任意两个边相等的三角形
    if a == b or a == c or b == c:
        # 等边：三个边都相等
        if a == b and a == c:
            print("这是一个等边三角形！")
        else:
            print("这是一个等腰三角形！")
    # 直角：a * a + b * b = c * c
    elif a * a + b * b == c * c or a * a + c * c == b * b or b * b + c * c == a * a:
        print("这是一个直角三角形！")
    # 其他的就是普通三角形
    else:
        print("这是一个普通的三角形！")
else:
    print("这三个边长不能构成三角形！！！")
```

```python
# 4.输入一个年份，判断是平年还是闰年
# 平年：年数除以4除不尽的叫做平年
# 闰年：能被4整除且不能被100整除，或者能够被400整除

year = int(input("请输入年份："))
if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print("该年份是闰年")
else:
    print("该年份是平年")
```

```python
# 5. 令faHeight为父亲的身高，moHeight为母亲的身高，childHeight为孩子的身高
# 男孩的身高预测公式：childHeight = (faHeight + moHeight) * 0.54  (cm)
# 女孩的身高预测公式：childHeight = (faHeight * 0.923 + moHeight) / 2 (cm)
# 是否喜欢体育运动，如果喜欢体育运动，那么就可以在基础身高上增加2%
# 是否有良好的卫生习惯，如果有良好的卫生习惯，那么就可以在基础身高上增加1.5%

# 要求：输入孩子父亲的身高，输入孩子母亲的身高，输入孩子的性别，输入是否喜欢体育运动和有没有良好的卫生习惯
# 通过以上的条件，预测出孩子的可能身高

# 获取父亲和母亲的身高数据
faHeight = int(input("请输入父亲的身高："))
moHeight = int(input("请输入母亲的身高："))

# 获取孩子的性别， boy表示男孩， girl表示女孩
sex = input("请输入孩子的性别：")

# 获取是否喜欢体育运动，用pe来表示，   yes or no
pe = input("请输入孩子是否喜欢体育运动：")

# 获取是否有良好的卫生习惯，用habit   yes or no
habit = input("请输入孩子是否有良好的卫生习惯：")

# 对孩子的性别做判断，根据不同的性别使用不同的身高预测公式
# 如果是男孩
if sex == 'boy':
    childHeight = (faHeight + moHeight) * 0.54

# 如果是女孩
elif sex == 'girl':
    childHeight = (faHeight * 0.923 + moHeight) / 2

pe_height = 0
habit_height = 0

# 判断是否喜欢体育运动，如果是yes，代表喜欢体育运动，那么就可以增加2%的身高
if pe == 'yes':
    pe_height = childHeight * 0.02

# 判断是否有良好的卫生习惯，如果是yes，代表有良好的卫生习惯，那么就可以增加1.5%的身高
if habit == 'yes':
    habit_height = childHeight * 0.015

# 最终的身高 = 基础身高 + 体育运动所带来的身高增益 + 良好的卫生习惯所带来的身高增益
finalHeight = childHeight + pe_height + habit_height

print(f'孩子的最终身高为:{finalHeight}')
```

```python
# 算术运算符
a = 3
b = 2

print("a + b的结果为：", a + b)
print("a + b的结果为：", a - b)
print("a + b的结果为：", a * b)
print("a + b的结果为：", a / b)
print("a + b的结果为：", a // b)
print("a + b的结果为：", a % b)
print("a + b的结果为：", a ** b)
```

```python
# 比较运算符
print(1 == 2)
print(1 != 2)
print(1 > 2)
print(1 < 2)
print(1 >= 2)
print(1 <= 2)
```

```python
# 定义变量 a 和 b
a = -10
b = 3

# 加法
print(a + b)

# 减法
print(a - b)

# 乘法
print(a * b)

# 除法
print(a / b)

# 整除
print(a // b)

# 取余
print(a % b)

# 幂运算
print(a ** b)
```

```python
a = -10
b = 3

# 计算商并调整商
quotient = a // b
if a % b != 0 and (a < 0) != (b < 0):
    quotient += 1

# 计算余数
remainder = a - quotient * b

print("商:", quotient)
print("余数:", remainder)
```
